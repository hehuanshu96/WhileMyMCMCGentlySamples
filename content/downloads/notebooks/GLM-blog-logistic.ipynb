# Bayesian Logistic Regression and The Oatmeal Posterior

Logistic Regression is commonly used when you have binary distributed data. It can also be used as a classifier.

The idea is pretty simple, we still estimate a linear model as before but pass the output (that ranges from $-\infty$ to $\infty$) through a logistic function which brings it into the the range [0, 1]. These 0-to-1 values can then be interpreted as the probabilities of a Binomial distribution. In other words, as we before had a Normal and T distribution, we now use a Binomial as our likelihood after passing the predictors through a link-function (the logistic).

Lets load some data about the height, weight and sex of individuals. One obvious hypothesis to test would be whether height or weight is a better predictor of sex (if you didn't know).

```python
import matplotlib.pyplot as plt
import numpy as np
from pymc import  *
import pandas as pd
from statsmodels.formula.api import glm as glm_sm
import statsmodels.api as sm
from pandas.tools.plotting import scatter_matrix
```

```python
htwt_data = pd.read_csv('data/HtWt.csv')
htwt_data.head(8)
```

As before, we can pass a `family` object to the `glm` function that specifies the likelihood distribution of our data. By default, these `families` also have a `link-function` associated with them. While by default the logistic function is used for `Binomial` you can also specify others (see `glm.links` for other options).

```python
with Model() as model_htwt:
    glm.glm('male ~ height + weight', htwt_data, family=glm.families.Binomial())
    trace_htwt = sample(1000, Slice(), progressbar=False)
```

In the models we estimated above we have been negligent about making sure that our estimation procedure (MCMC) actually worked (at least I didn't show this process but take my word that it worked well) and converged to the one-and-only true posterior. One way is to just plot the chains. Here we'll be looking at a scatter plot comparing pairwise all parameter combinations. Each dot is a sample generated from our MCMC algorithm while the diagonal contains the marginal of individual parameters.

```python
trace_df = trace_to_dataframe(trace_htwt)
scatter_matrix(trace_df, figsize=(8, 8));
```

_Posteriors are like Oatmeal -- if it's clumpy, it's no good._

This is a common problem with MCMC -- failure to converge. Fortunately, `PyMC3` supports not only Slice sampling but also *Hamiltonian Monte Carlo* (HMC) which is a new sampler that also takes the gradient of the posterior into account when choosing where to jump next and has much better convergence. NUTS is a recent extension that auto-tunes most parameters of HMC.

So lets estimate the same model but with `NUTS` instead of `Slice` by simply passing in another object to the `sample()` function.

```python
with Model() as model_htwt:
    glm.glm('male ~ height + weight', htwt_data, family=glm.families.Binomial())
    trace_htwt = sample(1000, NUTS(), progressbar=False)
```

```python
trace_df = trace_to_dataframe(trace_htwt)
scatter_matrix(trace_df, figsize=(8, 8));
```

There, nice and creamy -- yumm!

We can now also see why Slice sampling failed. There is huge correlation between the `height` coefficient and the `Intercept` which any Gibbs sampler like Slice is notoriously bad at (because it only updates one random variable each step). `HMC` however updates all random variables and is guided by the gradient which helps greatly with the correlation. Kruschke in his book *Doing Bayesian Data Analysis* re-parameterizes the model to still work with Gibbs sampling but this can be quite cumbersome.

Note that correlations are also a huge problem for frequentist GLMs that can cause the error to be underestimated and lead to wrong inference. Bayesian estimation on the other hand allows us to correctly model these correlations.

Finally, we can also do hypothesis testing on the posterior by looking at the probability mass below zero.

```python
print "P(weight < 0) = ", np.mean(trace_htwt['weight'] < 0)
print "P(height < 0) = ", np.mean(trace_htwt['height'] < 0)
```

We can thus say that height is highly predicative of sex.

Something I haven't talked about yet are *priors* which are another great benefit of the Bayesian framework. Specifically, we can easily build hierarchical GLMs in which individual coefficients share statistical power. I will leave this to a future blog post.

## Summary

 - `PyMC3` allows GLM specification with convenient syntax borrowed from R.
 - Different models can be expressed by changing the likelihood distribution and link function.
 - Advanced samplers like Hamiltonian Monte Carlo allow inference even when there are high correlations in the posterior.
